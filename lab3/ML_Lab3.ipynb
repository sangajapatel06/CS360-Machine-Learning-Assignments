```python
import pandas as pd
import numpy as np
data = pd.read_csv("student_scores.csv")

X = data['Hours']
Y = data['Scores']
import numpy as np
X = np.array(X)
Y = np.array(Y)*1.0
concat = []
for x in X:
  concat.append(x)
for y in Y:
  concat.append(y)
print("Concat length ",len(concat))
import statistics as stat
overallMean = stat.mean(concat)
overallMedian = stat.median(concat)
# overallMode = stat.mode(concat)
overallStdr = stat.stdev(concat)
print("Overall Mean ",overallMean)
print("Overall Median ",overallMedian)
print("Overall Stdev ",overallStdr)
```
